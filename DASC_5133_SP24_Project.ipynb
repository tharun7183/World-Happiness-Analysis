The report is a collaborative effort between the Sustainable Development Solutions Network (SDSN) and the Earth Institute at Columbia University, with contributions from various researchers and experts.

<br>

The report includes rankings of 158 countries based on the "World Happiness Index," which is calculated using survey data from the Gallup World Poll and other sources. The index combines factors such as GDP per capita, social support, healthy life expectancy, freedom to make life choices, generosity, and perceptions of corruption to assess overall happiness levels.

<br>

The 2015 Happiness Report sheds light on the relationship between happiness, well-being, and sustainable development, emphasizing the importance of incorporating measures of happiness into policy-making and development strategies. It provides valuable insights into global happiness levels, highlighting the factors that contribute to happiness and offering recommendations for policymakers and individuals to improve overall well-being.

```python
# create code block(s) as needed here
#Using KNN to build the data model:
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest
from sklearn import datasets, model_selection, metrics
x = df[["GDP","Social Support","Healthy Life Expectancy","Freedom","Corruption Perception","Generosity"]].astype(float)
y = df['Happiness Score'].astype(float) # Target variable (dependent variable)
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error


# Assuming X_train, y_train contain your training data

# List of k values to evaluate
k_values = list(range(1, 21))  # Example range, you can adjust this

# Dictionary to store mean cross-validation scores for each k
cv_scores = {}

# Perform 10-fold cross-validation for each k
for k in k_values:
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn_regressor, X_train, y_train, cv=10, scoring='neg_mean_squared_error')  # 10-fold CV for regression
    cv_scores[k] = -scores.mean()  # Take the negative mean squared error

# Find the k with the lowest cross-validation error
best_k = min(cv_scores, key=cv_scores.get)
print(f"Best k value: {best_k}")

# Create and train the KNN Regressor
knn_regressor = KNeighborsRegressor(n_neighbors=best_k)  # You can adjust n_neighbors
knn_regressor.fit(X_train, y_train)

# Predict on the test set
y_pred = knn_regressor.predict(X_test)

# Evaluate model performance (e.g., using Mean Squared Error)
mse = mean_squared_error(y_test, y_pred)
print(f"KNN Regressor Mean Squared Error: {mse}")
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Define the range of lambda values for regularization (alpha values for Ridge)
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}  # Adjust these values as needed

# Create the Ridge Regression model
ridge_reg = Ridge()

# Create Grid Search with cross-validation (cv=5 for example)
grid_search = GridSearchCV(ridge_reg, param_grid, cv=5, scoring='neg_mean_squared_error')

# Select the three best numerical features
x = df[["GDP", "Social Support", "Freedom"]].astype(float)
y = df['Happiness Score'].astype(float)  # Target variable (dependent variable)

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Fit the grid search to find the best lambda (alpha)
grid_search.fit(X_train, y_train)

# Get the best lambda (alpha)
best_alpha = grid_search.best_params_['alpha']

# Create a new Ridge Regression model using the best lambda
ridge_reg_best = Ridge(alpha=best_alpha)
ridge_reg_best.fit(X_train, y_train)

# Predict on the test set
y_pred_best = ridge_reg_best.predict(X_test)

# Evaluate model performance using Mean Squared Error (MSE)
mse_best = mean_squared_error(y_test, y_pred_best)
print(f"Ridge Regression Mean Squared Error with Best Alpha: {mse_best}")
print("The best regularization parameter alpha is : " + str(best_alpha))

#### Step #8: Use the model created using three features
# Predict the happiness score of three countries
country3 = pd.DataFrame([[1.1, 0.9, 0.9]], columns=["GDP", "Social Support", "Freedom"])
country4 = pd.DataFrame([[0.8, 0.7, 0.6]], columns=["GDP", "Social Support", "Freedom"])
country5 = pd.DataFrame([[0.9, 0.8, 0.7]], columns=["GDP", "Social Support", "Freedom"])

# Predict using the ridge regression model
country3_pred = ridge_reg_best.predict(country3)[0]
country4_pred = ridge_reg_best.predict(country4)[0]
country5_pred = ridge_reg_best.predict(country5)[0]

print("Predicted happiness score for Country 3:", country3_pred)
print("Predicted happiness score for Country 4:", country4_pred)
print("Predicted happiness score for Country 5:", country5_pred)

# Visualize the modeled relationship between Happiness Score and Social Support
plt.scatter(X_test['Social Support'], y_test, color='blue', label='Actual')
plt.scatter(X_test['Social Support'], y_pred_best, color='red', label='Predicted')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.title('Relationship between Happiness Score and Social Support')
plt.legend()
plt.show()

# Look at the coefficients and intercept
coefficients = ridge_reg_best.coef_
intercept = ridge_reg_best.intercept_
print("Coefficients:", coefficients)
print("Intercept:", intercept)
```
